```python
#UNIVARIATE ANALYSIS
#skewness above +1 or below -1
from scipy.stats import skew
skew(house_data['price'])
```

```python
#price is right skewed(positive)
house_data['price_norm'] = (house_data['price']-house_data['price'].mean())/house_data['price'].std()
```

```python
#taking the norm also didnt not remove skewness 
sns.distplot(house_data['price_norm'])
```

```python
#to remove postive skewneww we can use cubesqrt/sqrt/log
#of all the case in this scenario log give best in skewness so will use price_log 
from scipy.stats import skew
skew(house_data['price'])
house_data['price_cubesqrt']=house_data['price']**(1/2)
house_data['price_log']=np.log(house_data['price'])
```

```python
skew(house_data['price_cubesqrt'])
```

```python
skew(house_data['price_log'])
```

```python
#use age as years renovated and years built is  of no use
#use current year
house_data["house_age"] = 2018- np.where(house_data["yr_renovated"]==0,house_data["yr_built"],house_data["yr_renovated"])
```

```python
#calclating sqlft living and lot if they are changes for renovation
house_data["sqft_living"] = np.where(house_data["yr_renovated"]==0,house_data["sqft_living"],house_data["sqft_living15"])
house_data["sqft_lot"] = np.where(house_data["yr_renovated"]==0,house_data["sqft_lot"],house_data["sqft_lot15"])
print(skew(house_data["sqft_lot"]))
house_data['sqft_lot']=np.log(house_data['sqft_lot'])
print(skew(house_data["sqft_lot"]))
```
